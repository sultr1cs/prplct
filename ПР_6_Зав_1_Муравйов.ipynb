```python
import altair as alt
from vega_datasets import data
import pandas as pd
```

Завдання 1. Треба побудувати стовпчасту діаграму за допомогою бібліотеки Altair (Python).
Для цієї діаграми потрібно використовувати датасет source = data.wheat()

```python
source = data.wheat()
```

```python
data_chart = alt.Chart(source).mark_bar().encode(
    x='year:O',
    y="wheat:Q",
    color=alt.condition(
        alt.datum.year == 1810, #якщо рік буде 1810 то тест повертає TRUE,
        alt.value('orange'), #якщо буде НЕ TRUE то встановлюється блакитний колір
        alt.value('steelblue') #а якщо буде TRUE то встановиться помаранчовий колір
    )
).properties(width=600)

#Збереження файлу
data_chart.save('data_chart.html')

data_chart
```

Завдання 2. Треба побудувати стовпчасту діаграму допомогою бібліотеки Altair. Для цього завдання вам потрібно використовувати датасет source = data.wheat().


```python
bar = alt.Chart(source).mark_bar().encode(
    x='year:O',
    y='wheat:Q'
)

rule = alt.Chart(source).mark_rule(color='red').encode(
    y='mean(wheat):Q'
)

#Збереження файлу
(bar + rule).save('bar+rule.html')

(bar + rule).properties(width=600, title='Виробництво пшениці за роками')
```

Завдання 3. Треба Побудувати інтерактивний графік, використовуючи бібліотеку Altair . Використовувати датасет source = data.stocks()

```python

stocks = data.stocks()
source = stocks.groupby([pd.Grouper(key="date", freq="6M"), "symbol"]).mean().reset_index()

mybump = alt.Chart(source).mark_line(point=True).encode(
    x=alt.X("date:O").timeUnit("yearmonth").title("date"),
    y="rank:O",
    color=alt.Color("symbol:N")
).transform_window(
    rank="rank()",
    sort=[alt.SortField("price", order="descending")],
    groupby=["date"]
).properties(
    title="Bump Chart for Stock Prices",
    width=600,
    height=150,
)

#Збереження файла
mybump.save("bump_chart.html")  

mybump
```
