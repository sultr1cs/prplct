```python
import pandas as pd
import matplotlib.pyplot as plt
file = pd.read_csv('spotify-2023.csv', encoding = 'latin1')
file
```

```python
#task1
top_10 = file[file.released_year == 2023].sort_values('streams', ascending = False).head(10)

plt.figure(figsize=(15, 10))
plt.barh(top_10['track_name'], top_10['streams'], color='purple')
plt.xlabel('кількість прослуховувань', fontsize = 25)
plt.ylabel('назва пісні', fontsize = 25)
plt.title('10 найпопулярніших пісень у 2023 році за кількістю прослуховувань', fontsize = 30)
plt.show()
```

```python
#task2
import pandas as pd
import matplotlib.pyplot as plt

top_10 = file[file.released_year == 2023].sort_values('streams', ascending=False).head(10)
track_names = top_10['track_name']
danceability_values = top_10['danceability_%']

plt.figure(figsize=(8, 8))


plt.pie(danceability_values, labels=track_names, autopct='%1.1f%%', startangle=140)

plt.axis('equal')  
plt.title('Розподіл Danceability серед найпопулярніших пісень у 2023 році')

plt.show()
```

```python
#task3

# Створення невеликого датасету для прикладу бо в нашому датасеті немає потрібних даних
data = {
    'Дата': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01'],
    'Кількість прослуховань': [10000, 12000, 15000, 11000, 13000]
}

#створення датафрейму з цими даними
df = pd.DataFrame(data)
df['Дата'] = pd.to_datetime(df['Дата'])

#cтворення нашого лінійного графіку
plt.figure(figsize=(10, 6))
plt.plot(df['Дата'], df['Кількість прослуховань'], marker='o')
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кількість прослуховань', fontsize=12)
plt.title('Зміни кількості прослуховань протягом 2023 року', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
