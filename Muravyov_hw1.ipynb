```python
#5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими
#числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення.
array1 = np.random.randint(1, 21, 6)
array2 = np.random.randint(1, 21, 6)
print("Масив 1:", array1)
print("Масив 2:", array2)
add = array1 + array2
diff = array1 - array2
mult = array1 * array2

print('додавання:',add )
print('віднімання:', diff )
print('множення:', mult )
```


# Завдання 2. Робота з бібліотекою Pandas

```python
import pandas as pd

url = "https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

tables = pd.read_html(url)
table = tables[20]

table.head(10)
```

```python
#2. Визначити кількість рядків і стовпців в датафрейме (метод shape)
tables = pd.read_html(url)
table = tables[20]  

num_rows, num_columns = table.shape

print(f"Кількість рядків: {num_rows}")
print(f"Кількість стовпців: {num_columns}")
```

```python
#3. Замініть значення "—" в таблиці значеннями NaN
tables = pd.read_html(url)
table = tables[20]  

table = table.replace("—", np.nan)
print(table)
```

```python
#4. Визначте типи всіх стовпців за допомогою dataframe.dtypes
tables = pd.read_html(url)
table = tables[20]  

column_types = table.dtypes
print(column_types)
```

```python
#5. Замініть типи нечислових стовпців числовими стовпцями. Це стовпці, у яких було розташовано символ "—"
tables = pd.read_html(url)
table = tables[20]  

table = table.replace("—", np.nan).apply(pd.to_numeric, errors='coerce')

print(table)
```

```python
#6. Обчислити, скільки проміжків у кожному стовпці (використовуйте методи isnull, sum та len)
tables = pd.read_html(url)
table = tables[20]  

missing_values = table.isnull().sum()
print(missing_values)
```

```python
#7. Видалити дані для всієї країни з таблиці в останньому рядку таблиці
tables = pd.read_html(url)
table = tables[20]  

table = table.drop(table.index[-1])
print(table)
```

```python
#8. Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком (метод fillna)
pd.set_option('display.float_format', '{:.0f}'.format)
tables = pd.read_html(url)
table = tables[20]  

table = table.fillna(table.mean())
print(table)
```

```python
#9. Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні
tables = pd.read_html(url)
table = tables[20]  

avarage = table['2014'].mean()
highest = table[table['2014'] > avarage]
list_of_regions = highest['регіон'].tolist()

print(list_of_regions)
```

```python
#10. В якому регіоні була найвища народжуваність у 2010 та 2012 роках (окремо).
tables = pd.read_html(url)
table = tables[20]  
# я зробив з 2008 бо в таблиці немає стовпця з 2010 роком
data_2008 = table[['регіон', '2008']]
highest_birth_rate_2008 = data_2008[data_2008['2008'] == data_2008['2008'].max()]

print("Найвища народжуваність у 2008 році була в регіоні: \n", highest_birth_rate_2008)

data_2012 = table[['регіон', '2012']]
highest_birth_rate_2012 = data_2012[data_2012['2012'] == data_2012['2012'].max()]

print("\nНайвища народжуваність у 2012 році була в регіоні:\n", highest_birth_rate_2012)
```

```python
#11. Побудуйте стовпчасту народжуваності за регіонами у 2014 році
import matplotlib.pyplot as plt

tables = pd.read_html(url)
table = tables[20]  

data_2014 = table[['регіон', '2014']]
data_2014['2014'] = pd.to_numeric(data_2014['2014'], errors='coerce')
data_2014 = data_2014.dropna()

plt.figure(figsize=(12, 6))
plt.bar(data_2014['регіон'], data_2014['2014'])
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2014 році')
plt.title('Народжуваність за регіонами у 2014 році')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
