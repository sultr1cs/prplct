```python
#task 4
data = pd.read_csv('bestsellers with categories.csv', encoding = 'latin1')
data.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
df = pd.DataFrame(data)
df
```

```python
#task 5
missing_values = data.isna().sum()
print(missing_values)
print("\nПропусків немає")
```

```python
#task 6
unique_genres = data["genre"].unique()

print('є два унікальних жанри :', unique_genres)
```

```python
#task 7
max_price = data['price'].max()
min_price = data['price'].min()
mean_price = data['price'].mean()
median_price = data['price'].median()

print(f"Максимальна ціна: |{max_price}| \nМінімальна ціна: |{min_price}| \nСередня ціна: |{mean_price}| \nМедіанна ціна: |{median_price}|")
```

```python
#task 8
#Який рейтинг у датасеті найвищий?
max_rating = data['user_rating'].max()
max_rating_book = data[data['user_rating'] == max_rating]

#Скільки книг мають такий рейтинг?

count_max_rating_books = len(max_rating_book)

#Яка книга має найбільше відгуків?

max_reviews = data.loc[data['reviews'].idxmax()]
name_max_reviews_book = max_reviews['name']

#З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ?

year_2010 = data[(data['year'] == 2010)]
max_price = year_2010['price'].max()
most_expensive_book = year_2010[year_2010['price'] == max_price]

#Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році

fiction_genre = data[(data["genre"] == "Fiction")]
date = fiction_genre[(fiction_genre['year'] == 2010)]
book_count = len(date)

# Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)?

high_rating_2010_2011 = data[(data['user_rating'] == 4.9) & (data['year'].isin([2010, 2011]))]
book_counter = len(high_rating_2010_2011)

print(f"Максимальний рейтинг в списку: {max_rating} \n\nКількість книг з найвищим рейтингом: {count_max_rating_books} \n\nКнига з найбільшою кількістю відгуків: {name_max_reviews_book}")
print(f"\nЗ тих книг, що потрапили до Топ-50 у 2010 році, найдорожча книга: \n{most_expensive_book['name'].iloc[0]}")
print(f"\nКількість книг жанру 'Fiction', які потрапили до Топ-50 у 2012 році: {book_count}")
print(f"\nКількість книг з рейтингом 4.9, які потрапили до Топ-50 у 2010 або 2011 роках: {book_counter}")
```

```python
#task 9
filtered_data = data[data['genre'].isin(['Fiction', 'Non Fiction'])]

aggregations = {
    'price': ['min', 'max']
}

result = filtered_data.groupby('genre').agg(aggregations)
print(result)
```
